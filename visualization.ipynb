```python
import matplotlib.pyplot as plt
import pandas as pd
 
# Reading the tips.csv file
data = pd.read_csv(r"C:\Users\manas\Documents\Task\Answers\sponsors\sponsors.csv")
 
# initializing the data
x = data['facebook_id']
y = data['sponsor_name']
 
fig = plt.figure(figsize = (25, 5))
    
# plotting the data
plt.bar(x, y,linewidth=5)
 
# Adding title to the plot
plt.title("sponsors_dataset")
 
# Adding label on the y-axis
plt.ylabel('facebook_id')
 
# Adding label on the x-axis
plt.xlabel('Day')
 
plt.show()
```
